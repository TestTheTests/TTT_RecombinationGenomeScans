# Plots of Summary Values for Each Statistic

These plots are used to explore the starting data and test different values for the number of windows. The
CNN takes a feature vector derived by splitting each chromosome into 11 windows, labeling each window, taking
the mean of each statistic over the window, and dividing that mean value by the total of all mean values of the
statistic over the window. The code below calculates these scaled summary values for an input statistics file over
a given number of windows and plots them against their position in the genome.

```python
from plotnine import *
import pandas as pd
from featureVecUtils import *

fileDir  = "/media/kevin/TOSHIBA_EXT/TTT_RecombinationGenomeScans/results_final/"
cmd      = 'ls ' + fileDir + '*ScanResults.txt'
fileList =  subprocess.run(cmd, shell = True, stdout=subprocess.PIPE).stdout.decode('utf-8')
fileList = fileList.split("\n")
fileList = list(filter(None, fileList))

metricIndices   = [1,2,8,9,11,12,13,14,19,22,24,25,29]
file            = fileList[2]
features        = readFeatures(file, metricIndices)
statsOfInterest = ['pos', 'Hscan_v1.3_H12', 'pca_ALL_PC1_loadings', 'pca_ALL_PC2_loadings',
                   'pca_ALL_PC3_loadings', 'pcadapt_3.0.4_PRUNED_log10p', 
                   'OutFLANK_0.2_ALL_log10p', 'LFMM_ridge_0.0_ALL_log10p', 
                   'LFMM_lasso_0.0_ALL_log10p', 'Spearmans_ALL_rho']

featMask  = features['keep_loci'] == "TRUE"
goodFeats = features[featMask]

dictSplitChr = {chrom: df for chrom, df in goodFeats.groupby('chrom')}
#### We now have a dictionary where the keys are the chromosome numbers and the values are
#### a pandas dataframe of all snps on that chromosome
```

```python
def writeDf(nwin):
    first = True
    dfLabels = list(statsOfInterest)
    dfLabels.remove('pos')
    for key in dictSplitChr:
        summary = summarizeByWindow(dictSplitChr[key][statsOfInterest], nwin = nwin, scaled = True)
        if first:
            summariesDf = pd.DataFrame(summary, columns = dfLabels)
            first = False
        else:
            summaryDf   = pd.DataFrame(summary, columns = dfLabels)
            summariesDf = summariesDf.append(summaryDf, ignore_index=True)

    genomeEnd = 500000
    nchrom    = 10
    start     = genomeEnd / nwin / nchrom / 2
    
    genomePos = np.linspace(start, genomeEnd, nchrom * nwin, endpoint = False)
    summariesDf['location'] = genomePos
    
    return summariesDf
```

```python
def resetPlot(summariesDf, ymax = 1):
    summariesDf['chr5xmin'] = 220000
    summariesDf['chr5xmax'] = 230000
    summariesDf['ymin'] = 0
    summariesDf['ymax'] = ymax
    summariesDf['chr6xmin'] = 270000
    summariesDf['chr6xmax'] = 280000
    summariesDf['chr7xmin'] = 320000
    summariesDf['chr7xmax'] = 330000
    summariesDf['chr8xmin'] = 370000
    summariesDf['chr8xmax'] = 380000
    summariesDf['chr9xmin'] = 400000
    summariesDf['chr9xmax'] = 450000

    #plot = ggplot(aes(x='location'), data = summariesDf) #+ geom_line()
    plot = ggplot(data = summariesDf) #+ geom_line()
    
    #### Add rectangles for genomic features
    plot = plot + geom_rect(aes(xmin="chr5xmin", xmax="chr5xmax",ymin="ymin",ymax="ymax"), 
                            color="grey", alpha=0.005, fill='lightgrey')
    plot = plot + geom_rect(aes(xmin="chr6xmin", xmax="chr6xmax",ymin="ymin",ymax="ymax"), 
                            color="gray", alpha=0.005, fill='lightgray')
    plot = plot + geom_rect(aes(xmin="chr7xmin", xmax="chr7xmax",ymin="ymin",ymax="ymax"), 
                            color="red", alpha=0.005, fill='red')
    plot = plot + geom_rect(aes(xmin="chr8xmin", xmax="chr8xmax",ymin="ymin",ymax="ymax"), 
                            color="blue", alpha=0.005, fill='blue')
    plot = plot + geom_rect(aes(xmin="chr9xmin", xmax="chr9xmax",ymin="ymin",ymax="ymax"), 
                            color="lightblue", alpha=0.005, fill='lightblue')
    #### Add lines to delineate chromosomes
    plot = plot + geom_vline(xintercept = range(50000, 550000, 50000), linetype="dotted")
    
    return plot
```

```python
#### plot with nwindows = 11
df11win = writeDf(11)

plot = resetPlot(df11win, .25)
for stat in statsOfInterest[1:]:
    p = plot + geom_point(aes(x='location', y = stat)) + ggtitle(stat)
    print(p)
```

```python
def plotPreds(preds):
    nchrom  = 10
    predsDf = pd.read_csv(preds, sep = '\s+') 

    ### pull out the relevant data from the file and put it in "long" format
    probCols = ["prob(bgSelection)", "prob(inversion)", "prob(lowRC)", "prob(neutral)", 
            "prob(QTL)", "prob(selectiveSweep)", "prob(variableRC)"]

    probabilities = predsDf[probCols[0]]
    for col in probCols[1:]:
        probabilities = probabilities.append(predsDf[col])
    
    classifiedWinStart = predsDf['classifiedWinStart']
    for i in range(1, len(probCols)):
        classifiedWinStart = classifiedWinStart.append(predsDf['classifiedWinStart'])

    classifiedWinEnd = predsDf['classifiedWinEnd']
    for i in range(1, len(probCols)):
        classifiedWinEnd = classifiedWinEnd.append(predsDf['classifiedWinEnd'])
    
    # get the predictions and put them in a list for labeling
    predictions = predsDf['predClass'].tolist()  
    
    classOfProb = []
    for col in probCols:
        classOfProb.extend([col] * nchrom)

    #### combine all the info into a dataframe
    plotDf = pd.DataFrame()
    halfWinWidth = 25000          # 50000 / 2, used to find the center
    plotDf['classifiedWinStart'] = classifiedWinStart
    plotDf['classifiedWinEnd']   = classifiedWinEnd
    plotDf['classifiedWinCenter']= plotDf['classifiedWinStart'] + halfWinWidth
    plotDf['probability']        = probabilities
    plotDf['class']              = classOfProb
    
    labelsX = plotDf['classifiedWinCenter']
    labelsX = labelsX.iloc[0:10]
    labelsDf = pd.DataFrame(
        {'x'         : labelsX,
        'y'          : [0.5] * 10,
        'prediction' : predictions
        })
  
    plot = ggplot(data=plotDf) + geom_bar(aes(x = 'classifiedWinCenter', y = 'probability',
                                              fill = 'class', width = 48000),
                                          stat="identity") + ggtitle("Predictions") \
            + labs(x = "Genomic Position") \
            + geom_text(aes(x = 'x', y = 'y', label = 'prediction'), data = labelsDf, angle = 90)
    
    print(plot)

    ##### Now add a column with the true labels (prob = 1 for actual class, 0 for all others)
    BGS = ([ 0 ] * 5)
    BGS.extend([1,1,0,0,0])

    inv = ([ 0 ] * 7)
    inv.extend([1, 0, 0])

    lowRC = ([ 0 ] * 8)
    lowRC.extend([1, 0])

    neut  = [1,1]
    neut.extend([ 0 ] * 8)

    qtl   = [0,0]
    qtl.extend([1,1])
    qtl.extend([ 0 ] * 6)

    ss    = [ 0 ] * 4
    ss.append(1)
    ss.extend([ 0 ] * 5)

    varRC = [0] * 9
    varRC.append(1)

    trueProb = BGS + inv + lowRC + neut + qtl + ss + varRC  # append it all into one column
    plotDf["trueProb"] = trueProb                           # store it in the dataframe  
    
    # x and y are the same as the first graph, change the labels to be the true labels instead of predictions
    labelsDf['trueLabel'] = ["Neutral", "Neutral", "QTL", "QTL", "SS",
                             "BGS", "BGS", "Inversion", "lowRC", "varibleRC"]
    
    truePlot = ggplot(data=plotDf) + geom_bar(aes(x = 'classifiedWinCenter', y = 'trueProb',
                                              fill = 'class', width = 48000),
                                          stat="identity") + ggtitle("True Class")\
            + labs (x = "Genomic Position", y = "Probability")\
            + geom_text(aes(x = 'x', y = 'y', label = 'trueLabel'), data = labelsDf, angle = 90)
    print(truePlot)
```
